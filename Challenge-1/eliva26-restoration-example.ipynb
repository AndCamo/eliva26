```python
#encode an image to embed it in the .csv file
def encode_img(img):
    # Encode image into base64 text
    retval, buffer = cv2.imencode('.jpg', img)
    return base64.b64encode(buffer).decode('utf-8')
```

```python
# modify this function to apply your solution to each image
def restore(img):
    img_e = cv2.GaussianBlur(img, (0,0 ),sigmaX=5,sigmaY=5)
    return img_e
```

```python
#loop over the images in data folder
# apply the enhancemnt 
# store the results in submission.csv file
# The obtained .csv can be submitted to the challenge

solution = {'id': [], 'data': []}
for path in glob('./data/*.jpg'):
    img = cv2.imread(path)
    # do image restoration
    enhanced = restore(img)

    # encoder_img to solution.csv
    solution['id'].append(os.path.basename(path))
    solution['data'].append(encode_img(enhanced))

df = pd.DataFrame(solution)
df.to_csv('submission.csv', index=False)
```
